## Loading out saved model in a fresh notebook

```python
from previous_chapter_four import GPTModel
import torch
```

```python
GPT_CONFIG_124M = {
  "vocab_size": 50257,
  "context_length": 256,
  "embed_dim": 768,
  "n_heads": 12,
  "n_layers": 12,
  "drop_rate": 0.1,
  "qkv_bias": False
}

model = GPTModel(GPT_CONFIG_124M)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.mps.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
model_state_dict = torch.load("model.pth", map_location=device)
print(f"Loaded model state dict on {device} with keys: {model_state_dict.keys()}")
model.load_state_dict(model_state_dict)
```
